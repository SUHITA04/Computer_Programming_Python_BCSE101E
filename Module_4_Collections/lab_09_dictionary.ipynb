# Dictionary
- a key-value pair.
- used to store data values like a map (hashing). 

**Dictionary in Python** is a collection of keys values, used to store data values like a map, which, 
unlike other data types which hold only a single value as an element.

### Properties of Keys and Values of dictionary
**Keys**:
- keys in a dictionary can not be repeated.
- keys must be immutable. 

**Values**
- Values in a dictionary can be of any data type.
- Values can be duplicated

## Initialize/Create a dictionary with following entry
{1: 'Python', 2: 'Java', 3: 'Ruby', 4: 'Scala'}

```python
d= { 8 : 'cppp',
    3: 'Python', 2: 'Java', 1: 'Ruby',
    11: 'Cplus',
    4: 'Scala'}
```

```python
# sorting the keys_list
keys_list = list(d.keys())
print(keys_list)
#keys_list.sort()
#print(keys_list)
```

```python
# sorting the values list
values_list = list(d.values())
print(values_list)
values_list.sort()
print(values_list)
```

```python
# custom sorting of the values_list based inline function - `lambda`
values_list.sort(key=lambda item: len(item))
print(values_list)
```

```python
# user defined function
def my_add(x, y):
  z = x+y
  return z
```

```python
# calling of user defined function my_add
t = my_add(2,3)
print(t)
```

```python
# sorting the dictionary items
items_list = list(d.items())
print(items_list)
items_list.sort()
print(items_list)
```

```python
# custom sort on the values of dictionary items
items_list.sort(key = lambda item : len(item[1]), reverse=True)
print(items_list)
```

## iterating over dictionary

```python
print(d)
for key, value in d.items():
  print(f'key={key}, value={value}')
```

```python
for key in d.keys():
  print(f'key={key}, value={d[key]}')
```

```python
# accessing elements of dictionary using sorted keys
for key in sorted(d.keys()):
  print(f'key={key}, value={d[key]}')

## What is the difference between sort and sorted ?
```
