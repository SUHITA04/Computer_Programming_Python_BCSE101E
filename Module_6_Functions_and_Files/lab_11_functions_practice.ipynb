- To find factorial of an interger using iterative and recursive approach

```python
def recursive_fact(n):
  """ calculate factorial of an positive integer using recusrion
  """
  if n <= 1:
    return 1
  else:
    return n* recursive_fact(n-1)
  
```

```python
def iterative_fact(n):
  """
    calculate factorial of an positive integer using iteration
  """
  fact =1 
  for num in range(2, n+1):
    fact = fact * num
  return fact
```

```python
x_list = [ num  for num  in range(15)]
print(x_list)
y_list = [ recursive_fact(num) for num in x_list]
print(y_list)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_list,y_list)
```

## to find fibbonacci series $n^{th}$ number

```python
def fibbonacci(n):
  if n <=1:
    return n
  else:
    return fibbonacci(n-1) + fibbonacci(n-2)
```

```python
x_list = [ num for num in range(10)]
y_list = [ fibbonacci(num) for num in x_list]
print(x_list)
print(y_list)
```

```python
n = int(input('Enter a number: '))
print(fibbonacci(n))
```

## To find GCD of a number

```python
def gcd(num1, num2):
  """
  Euclidian algorithm using substraction
  """
  if num1 == num2:
    return num1
  elif num1 > num2:
    return gcd(num1-num2, num2)
  else:
    return gcd(num1, num2 - num1)
```

```python
print(gcd(16,24))
```

```python
num1 = int(input('Enter a number: '))
num2 = int(input('Enter a number: '))
num_gcd  = gcd(num1, num2)
print(f'num1= {num1}, num2 = {num2}; gcd={num_gcd}')
```

```python
def gcd2(num1, num2):
  if num2 == 0:
    return num1
  else:
    return gcd(num2, num1 % num2)
```

```python
num_gcd = gcd2(35,45)
print(num_gcd)
```

## LCM of two number

```python
def lcm(num1, num2):
  num_gcd = gcd(num1, num2)
  num_lcm = (num1*num2) / num_gcd
  return num_lcm
  
```

```python
num_lcm = lcm(12,18)
print(num_lcm)
```
